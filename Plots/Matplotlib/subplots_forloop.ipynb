```python
# Create a 3x3 subplot geometry in a for loop.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.random.randn(3,9)
n.flatten()
```

```python
fig, ax = plt.subplots(3,3, figsize=(7,7))
for i in ax.flatten():
  i.plot(np.random.randn(3,4))

plt.tight_layout()
plt.show()
```
