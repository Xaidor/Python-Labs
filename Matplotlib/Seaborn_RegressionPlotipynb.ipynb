```python
# Create a regression plot of data in matrix D -> df (data frame)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
n = 200
rggs = np.zeros((n,2))

rggs[:,0] = np.linspace(0,10,n) + np.random.randn(n)
rggs[:,1] = rggs[:,0]**2 + np.random.randn(n)*12

df = pd.DataFrame(data=rggs, columns=["var1","var2"])

sns.regplot(x=df.columns[0] ,y=df.columns[1] ,data=df ,color="purple")
plt.show()
```
