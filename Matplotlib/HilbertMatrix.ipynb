```python
# Visualize the Hilbert matrix using seaborn (heatmap)
# H is 10x10 matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from imageio import imread
```

```python
n = 10
H = np.zeros((n,n))

for i in range(n):
  for j in range(n):
    H[i,j] = 1/((i+1)*(j+1))

sns.heatmap(H)
plt.show()
```
